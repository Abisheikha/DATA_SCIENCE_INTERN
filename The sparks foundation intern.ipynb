##  Determining the optimum number of clusters using the elbow method

```python
from sklearn.cluster import KMeans
wcc=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',max_iter = 400 ,n_init=10,random_state=0)
    kmeans.fit(d)
    wcc.append(kmeans.inertia_)
    
mp.plot(range(1,11),wcc)
mp.title("elbow method")
mp.xlabel("Number of clusters")
mp.ylabel("Wcc")
mp.show()
```
